```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
tickdf = pd.read_html("https://en.wikipedia.org/wiki/Nasdaq-100")[4]
```

```python
tick=tickdf.Ticker.to_list()
```

```python
df=yf.download(tick,start = '2010-01-01')['Adj Close']
```

```python
df
```

```python
# observe the Nan values, for companies that were not in trade
# at the time, ex; TSLA, ABNB
```

```python
# to diminish survivorship bias, drop the Nan values
```

```python
df=df.dropna(axis=1)
```

```python
df
```

```python
# You can see that over 20 columns were dropped by dropping nan values
```

```python
# calculating monthly returns...
```

```python
mtly = (df.pct_change() +1)[1:].resample('M').prod()
# cumulating daily returns, resampling into monthly data, 
# and taking pproduct to obtain monthly pct change
# Also, first row is dropped as it becomes a Nan value as there
# is no pct change in the first value
```

```python
mtly
```

```python
# Creating a function to calclate rets for 3,6,12 months
def get_rolling_ret(df,n):
    return df.rolling(n).apply(np.prod)
```

```python
get_rolling_ret(mtly,12)
# This has rolling returns over 12 months, monthly, after the first 12 months
# As 12 rows are needed for the caluclation, 
# the first 12 rows are Nan values
```

```python
ret_12,ret_6,ret_3 = get_rolling_ret(mtly,12),get_rolling_ret(mtly,6),get_rolling_ret(mtly,3)
```

```python
ret_12
```

```python
ret_6
```

```python
ret_3
```

```python
# now having dataframes of returns from past 3,6,12 months
```

```python
top_50 = ret_12.loc['2010-12-31'].nlargest(50).index
# the top 50 performs for past 12 months at the end of 2010
```

```python
top_50
```

```python
top_30 = ret_6.loc['2010-12-31', top_50].nlargest(30).index
#top30 over the last 6 months, given that are top 50 over last 12 months                                                   
```

```python
top_30
```

```python
top_10 = ret_3.loc['2010-12-31',top_30].nlargest(10).index
# Once again now, filtering stocks with the top 10 3 month returns,
# given that they are top 30 in 6 month returns, given that they are 
# top 50 in 12 month returns
```

```python
top_10
```

```python
def get_top(date):
    top_50 = ret_12.loc[date].nlargest(50).index
    top_30 = ret_6.loc[date, top_50].nlargest(30).index
    top_10 = ret_3.loc[date,top_30].nlargest(10).index
    return top_10

# function to repeat above steps for any data
```

```python
get_top('2010-12-31')
```

```python
# now to view how the portfolio performance of those stocks would
# have been if it was held for a month
```

```python
mtly.loc['2010-12-31':][1:2]
# slicing the monthly return df from date that data was collected
# until the next month
```

```python
# getting the returns for the symbols that were identified
portfolio = mtly.loc['2010-12-31':,get_top('2010-12-31')][1:2]
portfolio
```

```python
portfolio.mean(axis=1).values[0]
# cumulative portfolio return on that date
```

```python
# creating a total function


def pf_performance(date):
    portfolio = mtly.loc[date:,get_top(date)][1:2]
    return portfolio.mean(axis=1).values[0]
```

```python
mtly.index
# any of these dates can be input into the performance function
```

```python
pf_performance('2011-10-31')-1
```

```python
# testing the strategy
```

```python
returns = [] 
for date in mtly.index[:-1]:
    returns.append(pf_performance(date))
    
    
# calculating performacne per month, for each month
# as the current month is incomplete, it is excluded
```

```python
returns
```

```python
print(pd.Series([i-0.01 for i in returns],index=mtly.index[1:]).cumprod().plot())
print(pd.Series([i-0.01 for i in returns],index=mtly.index[1:]).cumprod()[-1])

# transforming to more readable form
```

```python
# Full compiled function, which takes in a date and outputs the returns of the
# selected portfolio to day, as well as the selected portfolio
# entering 1 with the date uses the nasdaq, entering 2 uses the S&P500
```

```python
def Stock_Select_Backtest(date,num):
    if num%2==1:
        test_ticker_df = pd.read_html("https://en.wikipedia.org/wiki/Nasdaq-100")[4]
        test_tick = test_ticker_df.Ticker.to_list()
        test_df=yf.download(tick,start = '2010-01-01')['Adj Close']
        market = "Nasdaq 100"
        comp_df = yf.download('^IXIC', start = date)
#     chooses between creating the df with nasdaq or s%p
    else:
        test_ticker_df = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")[0]
        test_tick = test_ticker_df.Symbol.to_list()
        test_df=yf.download(tick,start = '2010-01-01')['Adj Close']
        market = "S%P 500"
        comp_df = yf.download('^GSPC', start = date)
    test_df=test_df.dropna(axis=1)
    test_mtly = (test_df.pct_change() +1)[1:].resample('M').prod()
    ret_12,ret_6,ret_3 = get_rolling_ret(test_mtly,12),get_rolling_ret(test_mtly,6),get_rolling_ret(test_mtly,3)
    print(get_top(date))
    test_returns = [] 
    for date in test_mtly.index[:-1]:
        test_returns.append(pf_performance(date))
    test_ret = pd.Series([i-0.01 for i in test_returns],index=test_mtly.index[1:]).cumprod()

    print(f"Cumulative returns from the selected top stocks in the {market} since {date}: {test_ret[-1]}")
    
    print(f"Graph of portfolio returns: {test_ret.plot()}")
    comp_ret = (comp_df['Adj Close'].pct_change()+1).cumprod()[-1]
    diff = ((test_ret[-1]-comp_ret)/(comp_ret))
    print(f"Since {date}, the {market} has returned {comp_ret}")
    print(f"The selected stocks had {diff*100}% better performance than the {market} did")
    
    
```
